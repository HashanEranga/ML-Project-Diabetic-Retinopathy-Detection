```python
targets_series = pd.Series(df_train['level'])
one_hot = pd.get_dummies(targets_series, sparse = True)
```

```python
i = 0 
for f, breed in tqdm(df_train.values):
    if type(cv2.imread('C:/Users/Hashan/Desktop/DL/Diabetic Retinopathy/data/resized_train/sample/{}.jpeg'.format(f)))==type(None):
        continue
    else:
        img = cv2.imread('C:/Users/Hashan/Desktop/DL/Diabetic Retinopathy/data/resized_train/sample/{}.jpeg'.format(f))
        label = one_hot_labels[i]
        x_train.append(cv2.resize(img, (im_size1, im_size2)))
        y_train.append(label)
        i += 1
np.save('x_train2',x_train)
np.save('y_train2',y_train)
print('Done')
```
